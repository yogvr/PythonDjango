# Proyecto del Tema

> El ahorcado. El juego es muy sencillo y popular, pero por si acaso te lo explico rápidamente. El programa va a elegir una palabra secreta y le va a mostrar al jugador solamente una serie de guiones que representa la cantidad de letras que tiene la palabra. El jugador en cada turno deberá elegir una letra y si la letra se encuentra en la palabra oculta, el sistema le va a mostrar en qué lugares se encuentra. Pero si el jugador dice una letra que no se encuentra en la palabra oculta, pierde una vida.

> En el juego real del ahorcado, cada vez que perdemos una vida, se va completando el dibujo del ahorcado miembro por miembro. Pero en nuestro caso, como aún no tenemos elementos gráficos, simplemente le vamos a decir que tiene seis vidas y se las iremos descontando de una en una, cada vez que el jugador elija una letra incorrecta.

> Si se agotan las vidas antes de adivinar la palabra, el jugador pierde. Pero si adivina la palabra completa antes de perder todas las vidas, el jugador gana. Parece sencillo, pero ¿cómo diseñamos todo este código? Bueno, aquí van algunas ayudas: 

* Primero vas a crear un código que importe el método choice, ya que lo vas a necesitar para que el sistema pueda elegir una palabra al azar dentro de una lista de palabras que también vas a crear al principio.
* Después de eso, vas a crear tantas funciones como creas necesarias para que el programa haga cosas como pedirle al usuario que elija una letra, para corroborar si lo que el usuario ha ingresado es una letra válida, para comprobar si la letra ingresada se encuentra en la palabra o no, para verificar si ha ganado o no, etc.
* Recuerda escribir primero las funciones y luego el código que las implementa ordenadamente.


```python
def ahorcado():
    words = "Ácido  Humilde  Planetario  Dispositivos  Gatear  Terapéutico  Valioso  Lujoso  Abrazar  Coincidencia  Intentar  Abordo  Minipíldora  Sin futuro  Biológico  Salvage  Conectividad  Empleado  Andrajoso  Zoológico Claro  Flood  Aviso  Pensamiento  Susurrar  Seguro   Encender  Hormiga  Ladrón  Norma  Inmolar  Título  Aplicar  Previo  Trauma  Divino  Saltamontes  Monótono  Auténtico Feroz  Prefabricado  Fin  Íntimo  Receta  Maldito  Despido  Morir  Temporal  Colapsar  Entero  Cardenal  Magnífico  Momento  Azar  Naturaleza  Fondo  Vomitar  Tiza  Arroyo Cautiverio  Agradable  Elfo  Carro  Drimys  Elegante  Estaca  Suponer  Cantidad  Tallar  Chop  Calentamiento  Engranaje  Niño  Tecnología  Joroba  Reino  Droga  Pecoso  Sucio Real  Abyecto  Usuario  Cómplice  Rodear  Ramita  Desagüe  Hobby  Abatido  Tontería  Mugre  Insalubre  Mantenimiento  Apocalíptico  Decir  Lluvia  Planta  Sate  Interesante  Trabajado Malhumorado  Bocado  Ajustable  Superar  Condición  Proponer  Invisible  Marcado  Participa  Rompecabezas  Desenfrenado  Abrasivo  Recomendación  Iluminar  Gracias  Nido  Palos Infinito  Lustre"
    abc = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
    monigote = ['''
    +---+
    |   |
        |
        |
        |
        |
    =========''', '''
    +---+
    |   |
    O   |
        |
        |
        |
    =========''', '''
    +---+
    |   |
    O   |
    |   |
        |
        |
    =========''', '''
    +---+
    |   |
    O   |
    /|   |
        |
        |
    =========''', '''
    +---+
    |   |
    O   |
    /|\  |
        |
        |
    =========''', '''
    +---+
    |   |
    O   |
    /|\  |
    /    |
        |
    =========''', '''
    +---+
    |   |
    O   |
    /|\  |
    / \  |
        |
    =========''']

    wordlist = list(set(words.split())) # elimino duplicados por si acaso
    abecedario = list(abc)

    print(f"{monigote[0]}\n> ¡Bienvenido al juego del ahorcado!\n> Voy a pensar una palabra y tendrás que adivinarla.\n> Tendrás un máximo de 6 intentos.")

    from random import choice

    palabra = choice(wordlist).upper()
    equivalencias = {"Á":"A", "É":"E","Í":"I","Ó":"O","Ú":"U","Ü":"U"}
    for contilde, sintilde in equivalencias.items():
        palabra = palabra.replace(contilde, sintilde).replace(contilde.lower(), sintilde)
    
    letras_palabra = list(palabra)
    palabraoculta = ["_" for i in letras_palabra if i in abecedario]
    vidasgastadas = 0
    letrasgastadas = set()

    print(f"> ¿Preparado? Vamos a jugar.\n > Mi palabra es: {palabraoculta}")
    
    
    while vidasgastadas < 6:
        letra = input("> Introduce una letra:\n\t").upper()

        if letra not in abecedario:
            print("> Creo que no me has entendido. Por favor, introduce una letra cualquiera:\n\t")

        elif letra not in letras_palabra:
            if letra not in letrasgastadas:
                letrasgastadas.add(letra)
                vidasgastadas += 1
                print(f"{monigote[vidasgastadas]}\n> No, esa letra no figura.\n> Hasta ahora has probado con las siguientes letras: {letrasgastadas}.\n\n> Mi palabra es: {palabraoculta}")
            else:
                 print("> Por favor, vamos a centrarnos. Ya me has dicho esa letra.\n> Prueba con otra.\n")
    
        elif letra in letras_palabra:
            if letra not in letrasgastadas:
                letrasgastadas.add(letra)

                while letra in letras_palabra: # con eso me aseguro de que todas las ocurrencias de la letra que ha dicho el jugador se desvelan, en vez de solo la primera (por usar index)
                    palabraoculta[letras_palabra.index(letra)] = letra
                    letras_palabra[letras_palabra.index(letra)] = 0

                    if all(i == 0 for i in letras_palabra) == True:
                        print(f"\n> ¡Has adivinado mi palabra!")
                        print(f"\t{palabra}")
                        return
                
                print(f"{monigote[vidasgastadas]}\n> Muy bien. Continúa.\n> Hasta ahora has probado con las siguientes letras: {letrasgastadas}.\n\n> Mi palabra es: {palabraoculta}\n")

            else:
                 print("> Por favor, vamos a centrarnos. Ya me has dicho esa letra.\n> Prueba con otra.\n")

    if vidasgastadas >= 6:
        print(f"{monigote[vidasgastadas]}\n¡Has perdido!\nMi palabra era {palabra}.")
    return

ahorcado()


```
